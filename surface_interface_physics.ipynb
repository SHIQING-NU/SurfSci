**A brief summary of plane groups**
- 5 Bravais lattices:</br>
There is only one centered lattice, oc. for monoclinic, orthorhombic, and hexagonal cases, the centered lattice reduces to primitive lattices with halved unit cells.
  - monoclinic primitive (mp)
  - orthorhombic primitive (op)
  - orthorhombic centered (oc)
  - tetragonal primitive (tp)
  - hexagonal primitive (hp)
- 17 plane groups: </br>
The symbols for plane groups, the Hermann-Mauguin symbol, are standard in crystallography. </br>
The first place: the type of lattice</br>
The second place: the axial symmetry (1-, 2-, 3-, 4-, and 6-fold)</br>
The letter $m$: a mirror reflection</br>
The letter $g$: a symmetry with respect to a glide line, one-half of the unit vector translation followed by a mirror reflection.</br>
The plane groups are not mutually unrelated. Some of them are subgroups of other plane groups.

The order of the factor group (the number of different symmetry operations)
- In monoclinic, orthorhombic, and tetragonal families: p1, p2, pm, pg, cm, p2mm, p2mg, p2gg, c2mm, p4, p4mm, p4gm;
- In the hexagonal family: p3, p3m1, p6, p31m, p6mm



**Invariant functions**

Any function with 2D periodicity can be expanded into a 2D Fourier series.

a function has additional symmetry other than translational $\rightarrow$ some of the terms in the Fourier expansion vanish & some non-vanishing Fourier coefficients equal each other $\rightarrow$ The number of independent parameters is reduced

the form of a quantity periodic in x and y would be:
$$Q(\bf{r})=Q_0(z)+Q_1(z)f_1(x, y)+Q_2(z)f_2(x, y)+...$$
where $f_1(x, y), f_2(x, y)$ are symmetrized sums of sinusoidal functions.</br>
Given distance $z$ $\rightarrow$ $Q_0(z)$ const.
  - _invariant function_: $f_1(x, y)$ etc.
normalized to be: $\text{max}[f_1(x, y)]-\text{min}[f_1(x, y)]=1$

  - corrugation amplitude: $Q_1(z)$ etc.

(1) Plane group pm:</br>
(110) plane of III-V semiconductors with op, like GaAs(110): a mirror reflection through a line connecting two Ga (or As) nuclei in the [001] direction.

The possible corrugation functions:

$$f_1(x,y)=\frac{1}{2}cos\frac{2n\pi x}{a}cos\frac{2m\pi y}{b}$$

$$f_2(x,y)=\frac{1}{2}sin\frac{2n\pi x}{a}cos\frac{2m\pi y}{b}$$

$$f_3(x,y)=\frac{1}{2}cos\frac{2n\pi x}{a}sin\frac{2m\pi y}{b}$$

$$f_4(x,y)=\frac{1}{2}sin\frac{2n\pi x}{a}sin\frac{2m\pi y}{b}$$

$m$ $\rightarrow$ $f_3$ and $f_4$ (X)

$f_1$ and $f_2$ have the same z dependence

$$f_{nm}(x, y)=\frac{1}{2}cos(\frac{2n\pi x}{a}+\phi_{nm})cos(\frac{2m\pi y}{b})$$
where $\phi_{nm}$ are phase constants

(2) Plane group p2gm:</br>

By replacing Ga and As atoms with the same species, like Si or Ge, the symmetry becomes higher. The additional gliding symmetry operation means that, by letting $y \rightarrow y+b/2$ and $x \rightarrow -x$ simultaneously, the func not change.
$$f_1(x, y) = \frac{1}{2}cos\frac{2n\pi x}{a}cos\frac{4m\pi y}{b}$$

$$f_2(x, y) = \frac{1}{2}sin\frac{2n\pi x}{a}cos\frac{(4m+2)\pi y}{b}$$

The general form of a corrugation function that contains the first four terms (in order of the absolute value of g):
$$Q(\bf{r})=Q_0(z)+ \frac{Q_1(z)}{2}cos\frac{2\pi x}{a}+\frac{Q_2}{2}cos\frac{4\pi x}{a}+\frac{Q_3(z)}{2}sin\frac{2\pi x}{a}cos\frac{2\pi y}{b}$$

The term $Q_0$ is a constant. The next two terms describe the corrugation in the [001] direction. The last term describes the corrugation in the [11 ̄10] direction.

(3) Plane group p2mm:</br>

Si(111)2x1, Si(100)2x1, etc. The symmetric function is similar to that of plane group p2gm.

(4) Plane group p4mm:</br>

By setting the origin of the coordinate system at the intersection of the two mirror reflection lines, only $f_1$ is invariant under the mirror reflection operation. 4-fold rotational symmetry $\rightarrow$ n = m, and a = b
$$Q(r)=Q_0(z)+\frac{1}{2}Q_1(z)cos\frac{2\pi x}{a}cos\frac{2\pi y}{a}$$
This function describes a simple tetragonal surface with atoms located at (na, ma), where n and m are arbitrary integers.

(5) Plane group p6mm:</br>

$$Q(r) = Q_0(z)+Q_1(z)\psi^{(6)}(kx)$$
where $k = \frac{4\pi}{\sqrt{3}a}$ is the length of a primitive reciprocal lattice vector, and 
$$\psi^{(6)}(X) \equiv \frac{1}{3}+\frac{2}{9}\sum_{n=0}^2cos\omega_n \cdot X$$ is the hexagonal cosine function, where $\omega_0=(0, 1), w_1=(-\frac{1}{2}\sqrt(3),-\frac{1}{2})$, and $w_2=(\frac{1}{2}\sqrt(3),-\frac{1}{2})$

**Surface stress/tension**

Surface: the charge distribution (details of the chemical bonds), the atomic scale forces on the atoms in the topmost layers are changed $\rightarrow$ $\tau_{ij}$ varies along an axis z perpendicular to the surface within several atomic distances

Surface stress tensor [force per unit length]:

$$\tau_{ij}^{(s)}=\int_{-\infit}^\infit dz[\tau_{ij}(z)-\tau_{ij}^{(b)}]$$

- the bulk $\tau_{ij}(z) = \tau_{ij}^{(b)}$ $\rightarrow$ surface stress vanishes; it has non-vanishing values only within a thin film of several atomic layers.
- positive: tensile stress, contract (work needed to stretch the surface elastically)
- negative: compressive
- the order of 1 N/m
- confined to typically distances of 1 nm from the surface

**Surface free energy**

Work $\delta W$: straining a thin crystal plate of a thickness $t$ by a strain $\delta\varepsilon_{ij}$:

$$\delta W = A\int_{-t/2}^{t/2} dz \sum_{ij}\tau_{ij}(z)\delta \varepsilon_{ij}$$

Boundary conditions of constant temp and particle number: $\delta W = \delta F^{(s)}$

$$\delta W = \delta W^{(s)} + \delta W^{(b)}= \delta F^{(s)} + \delta F^{(b)}=2A\sum_{ij}\tau_{ij}^{(s)}\delta\varepsilon_{ij}+At\sum_{ij}\tau_{ij}^{(b)}\delta\varepsilon_{ij}$$

factor 2 in the first term is coz of the 2 surfaces

$\Rightarrow$ specific surface free energy $\gamma$

$$\delta F^{(s)}=\delta(\gamma A)=\gamma\delta A + A\delta\gamma$$

$\Rightarrow$ two contribution to $F^{(s)}$
- $\gamma\delta A$: the number of surface atoms is changed with a fixed average area per surface atom
- $A\delta\gamma$: energetics vary (the interatomic distances in a reconstruction change)

Consider $dA=A\sum_id\varepsilon_{ii}$

Shuttleworth equation:

$$\tau_{ij}^{(s)}=\gamma\delta_{ij}+\frac{\partial\gamma}{\partial\varepsilon_{ij}}$$

- liquid surface: $\gamma$ is identical to the isotropic surface stress $\tau$ (due to no resistance to plastic deformation) $\rightarrow$ second term vanishes

Consider $\partial\gamma/\partial\varepsilon=\tau^{(s)}-\gamma$ as a thermodynamic driving force to move atoms from the bulk into the surface layer
- $\tau^{(s)}-\gamma>0$: accumulate more atoms on surface
- $\tau^{(s)}-\gamma<0$: less atoms in the surface layer
In this sense, $\partial\gamma/\partial\varepsilon$ is also one driving force for the formation of certain surface reconstructions

## 2D Lattices, Superstructure, and Reciprocal Space

The surface region of a crystal is, in principle, a 3D entity; reconstructions usually extend into the crystal by more than one atomic layer.

Space-charge layers on semiconductor surfaces can have a depth of hundreds of Ångstroms

the experimental probes in surface experiments, even slow electrons, usually have a non-negligible penetration depth. But compared to subsurface layers, the topmost atomic layer is always predominant in any surface experiment.

$\Rightarrow$ each layer of atoms in the surface is intrinsically inequivalent to other layers, i.e. the only symmetry properties which the surface posesses are those which operate in a plane parallel to the surface. Although the surface region is 3D, all symmetry properties are 2D. Thus surface crystallography is 2D and one has to consider 2D point groups and 2D Bravais nets or lattices.




**Surface lattices**

Inversion centers, mirror planes, and rotation axes perpendicular to the surface.

point group operations: 1, 2, 3, 4, 6-fold rotation axes

$\Rightarrow$ 10 different point group symmetries denoted:

$$1, 2, 1m, 2mm, 3, 4m, 4, 4mm, 6, 6mm$$

The numeral $\nu=1...6$ denotes rotations by $2\pi/\nu$. The third m indicates that a combination of the preceding two operations produces a new mirror plane.

The operation of the 2D point groups on a 2D translational net or lattice $\Rightarrow$ possible 2D Bravais lattices

$\Rightarrow$ 5 symmetrically distinct nets (2D Bravais lattices)
- one non-primitive unit mesh, the centered rectangular one
- a centered unit cell described by a primitive Bravais lattice (centered square meshes for convenience)

Superlattice: a different periodicity is present in the topmost atomic layer, a surface lattice.

The basic substrate lattice can be described by a set of 2D translational vectors:

$$r_m=ma_1+na_2$$

The surface net of the topmost layer may then be determined in terms of the substrate net by:

$$\begin{matrix}b_1=m_{11}a_1+m_{12}a_2 \\ b_2=m_{21}a_1+m_{22}a_2 \end{matrix}$$

or

$$\left(\begin{matrix}b_1 \\ b_2\end{matrix}\right) = M\left(\begin{matrix}a_1 \\ a_2\end{matrix}\right)$$

For the areas B and A of the surface and of the substrate unit mesh:

$$B=|b_1\times b_2|=A\text{det}M$$

The determinant of M,

$$\text{det}M=\frac{|b_1\times b_2|}{|a_1\times a_2|}$$

- det M is an interger: simple superlattice
- det M is a rational number: coincidence lattice
- det M is an irrational number: incoherent lattice/incommensurate structure, where the adsorbate-substrate interaction is much less important than interactions between adsorbed particles itself $\rightarrow$ out of registry with the substrate net

Wood notation:
- on a certain substrate surface X{hkl} a reconstruction with ($b_1||a_1, b_2||a_2$) $b_1=pa_1, b_2=qa_2$
$$X\{hkl\}(p \times q) or X\{hkl\}c(p \times q)$$
- a rotation by a certain angle $R^\circ$:
$$X\{hkl\}(p \times q)-R^\circ$$

**2D Reciprocal lattice**

the translational vectors of the 2D reciprocal lattice $\bf{a}_i^*$:
$$a_1^*=2\pi\frac{a_2\times\hat{n}}{|a_1\times a_2|}, a_2^*=2\pi\frac{\hat{n}\times a_1}{|a_1 \times a_2|}$$

$$a_i^*\cdot a_j=2\pi\delta_{ij} \text{and} |a_i^*|=2\pi[a_i\sin\angle(a_ia_j)]^{-1}, i,j = 1,2$$

$\Rightarrow$ construct the reciprocal lattice geometrically to a given 2D network

A general translation vector in reciprocal space:

$$G_{hk}=ha_1^*+ka_2^*$$

the integer numbers h and k are Miller indices

the reciprocal network of a superstructure $(b_1^*,b_2^*)$ can be expressed in terms of the substrate reciprocal lattice $(a_1^*,a_2^*)$ by:

$$\begin{matrix}b_1^*=m_{11}^*a_1^*+m_{12}^*a_2^* \\ b_2^*=m_{21}^*a_1^*+m_{22}^*a_2^* \end{matrix}$$

or

$$\left(\begin{matrix}b_1^* \\ b_2^*\end{matrix}\right) = M\left(\begin{matrix}a_1^* \\ a_2^*\end{matrix}\right)$$

The matrices $M^*$ are related to one another by:

$$M^*=(M^{-1})^T$$

$(M^{-1})^T$ is the tanspose inverse matrix with:

$$m_{ii}=\frac{m_{ii}^*}{\text{det}M^*} \text{and} m_{ij}=\frac{-m_{ij}^*}{\text{det}M^*}$$



**2D Frenkel-Kontorova (FM) model**

All the atoms in the deeper layers are frozen at the fcc lattice sites. All top-layer atoms located on a substrate are modeled with a periodic potential, and all nearest neighboring atoms are connected by a spring force. system energy:

$$E=\sum_iV_S(r_i)+\sum_jk(l_j-b)^2/2$$

$$V_S(r)=\sum_GV_S(G)e^{-iG\cdot r}$$

- i: Au atoms
- j: bonds (springs) on the top layer
- k: spring constant
- $l_j$: length of the jth bond
- b: equilibrium bond length
- G: reciprocal vector of the lattice
- $V_S(G)$: the coefficient of the 2D Fourier expansion of the periodic substrate potential

Modify the substrate potential field by introducing a periodic sinusoidal deformation along the y direction:

$$V_S(r)=\sum_GV_S(G)e^{-iG[r-a\cdot\sin(\frac{2\pi r_x}{L_x})\cdot\hat{y}]}$$

- a: y-deformation amplitude
- $L_x$: periodic length of the herringbone pattern along the x direction
- $\hat{y}$: the unit vector along y direction

Each dislocation contains a certain amount of elastic energy in its surrounding strain field.

## Nucleation and Growth of Thin Films

**Modes of film growth**

Condensation of new material from the gas phase (molecular beam or gas phase ambient) is described by an impinging rate (number of particles per cm2 per second):
$$r = p (2\pi MkT_0)^{-1/2}$$
- p: vapor pressure
- M: molecular weight of the particles

Once a particle has condensed from the vapor phase, the various reaction paths:

- re-evaporate

- diffuse along the surface

adsorption, particularly at special sites like edges or other defects

overcome characteristic activation energies

Arrhenius-type exponential law $\rightarrow$ the number of particles being able to participate in the particular process. the desorption rate:
$$v\propto \exp(E_{\text{des}}/kT)$$
$E_{\text{des}}$: activation energy for desorption

- nucleation of more than one adsorbed particle

How many particles are needed to form a new nucleus, which further grows into an island?

- interdiffusion

decay and formation of 2D clusters must obey detailed balance. in equilibrium, there is no net growth of a film and so crystal growth must clearly be a non-equilibrium kinetic process.

$\Rightarrow$ The state which is obtained is not necessarily the most stable one, since it is kinetically determined. In general, certain parts of the overall process may be kinetically forbidden, whereas others may be in local thermodynamic equilibrium. In this case equilibrium arguments may be applied locally even though the whole growth process is a non-equilibrium process.

Monte Carlo computer simulations

Film growth phenomenologically:
- layer-by-layer growth mode (or Frank-van der Merve, FM)
- island growth (or Vollmer-Weber, VW): multilayer conglomerate of adsorbed atoms.
- layer-plus-island growth mode (or Stransky-Krastanov, SK)

Since $\gamma$ can also be interpreted as a force per unit length of boundary, force equilibrium at a point where substrate and a 3D island of the deposited film touch:
$$\gamma_S=\gamma_{S/F}+\gamma_F\cos\phi$$

layer-by-layer (FM) and island (VW), can be distinguished by the angle $\phi$:
- layer growth: $\phi=0$, $\gamma_S\geq \gamma_F+\gamma_{S/F}$
- island growth: $\phi>0$, $\gamma_S < \gamma_F+\gamma_{S/F}$

The lattice of the film tries to adjust to the substrate lattice, but at the expense of elastic deformation energy. The transition from layer to island growth occurs when the spatial extent of the elastic strain field exceeds the range of the adhesion forces within the deposited material.

Consider the equilibrium condition for the whole system including the gas phase above the deposited film:
- f this transfer occurs exactly at the equilibrium vapor pressure p0(T ), then no energy is needed because of the equilibrium condition μsolid p0(T ) = μvapor( p0, T ).
- If however the particle changes over from vapor to solid at a pressure p, a free enthalpy change: 
$$\Delta G = n\Delta\mu=nkT\ln(p/p_0)$$

the degree of supersaturation: $\zeta=p/p_0$, is one of the “driving forces” for the formation of a thin film deposited from an ambient vapor phase

Taking this vapor phase into account, the conditions for layer or island growth have to be supplemented in the following way (C is a constant):
- layer growth: $\gamma_S \geq \gamma_F+\gamma_{S/F}+CkT\ln(p_0/p)$
- island growth: $\gamma_S < \gamma_F+\gamma_{S/F}++CkT\ln(p_0/p)$

$\Rightarrow$ the growth mode can be changed by varying the supersaturation conditions. With increasing supersaturation, layer-by-layer growth is favored.
- The level of supersaturation determines how atoms condense and form thin films.
- Very high $\zeta$ $\rightarrow$ rapid nucleation, possible amorphous or nanocrystalline films.
- Low $\zeta$ $\rightarrow$ controlled epitaxial growth with fewer defects.
- Higher $p_0$ due to heated substrate (esp. for group V elements)

**Capillary model of nucleation $\rightarrow$ predict general trends in nucleation and film growth**

describe island (3D cluster) and layer-by-layer (2D cluster) growth of nuclei on an ideal, defect-free, substrate surface (by Bauer)

$\Delta G <0$

3D nucleation: $\Delta G_{3D}=jkT\ln(p_0/p)+j^{2/3}X=-j\Delta\mu+j^{2/3}X$

j: the number of atoms forming the nucleus

quantity X contains the contributions of the interface energy

$$X=\sum_kC_k\gamma_F^{(k)}+C_{S/F}(\gamma_{S/F}-\gamma_S)$$

$C_{S/F}$: a simple geometrical constant which relates the basis area of the nucleus $A_{S/F}$ with the number of atoms according to $A_{S/F}=C_{S/F}j^{2/3}$

$C_k$ relates $j^{2/3}$ to a part of the surface of the nucleus (adjacent to vacuum) having the surface energy $\gamma_{F}^(k)$

The outer surface of the nucleus, i.e the part exposed to the vapor phase (or vacuum) is assumed to be composed of several patches of different crystallographic orientation with different surface energies $\gamma_F^{(k)}$

hemispherical nucleas with radius r, $\Omega$ as atomic volume of the film material:

$$\Delta_{3D}=-\frac{1}{2}\frac{4}{3}\pi r^3\Omega^{-1}\Delta\mu+2\pi r^2\gamma_F+\pi r^2(\gamma_{S/F}-\gamma_S)$$

$C_{S/F}$ would amount to $\pi^{1/3}(3\Omega/2)^{2/3}$

edge energy $\gamma_E$ [per unit length]: the amount of energy necessary to position additional film atoms along the unit length of such an edge.

The free enthalpy for 2D cluster growth:

$$\Delta_{2D}=-j\Delta\mu+j(\gamma_F+\gamma_{S/F}-\gamma_S)\Omega^{2/3}+j^{1/2}Y$$

$Y$: effect of the adsorbate (film) atoms at the edges

$$Y=\sum_lC_{l}\gamma_E^{(l)}$$

Sum over $l$: crystallographically different edge orientation $l$ may be associated with different edge energies $\gamma_E^l$. 

$C_l$ are geometrical factors defined as the $C_k$ and $C_{S/F}$

$\Rightarrow$ the last terms are positive $\rightarrow$ an increase of the free enthalpy with growing size of the nuclei $\rightarrow$ nucleation barrier

$\Rightarrow$ the first terms ($\propto \Delta\mu$) $\propto$ the logarithm of the $\zeta$ are negative $\rightarrow$ drive the nucleation process faster with increasing supersaturation and growing nucleus size

$\Rightarrow$ the superposition of the two effects $\rightarrow$ non-monotonic dependence of $\Delta G$ on the number of atoms $j$ within the nucleus. a critical size of the nuclei both for nucleation ($\Delta G$ becomes a maximum). Below this critical cluster size, the nucleus is not stable, it tends to decay because $\Delta G$ increases with growing size.

On the assumption of a circular, planar nucleus of radius r formed by one atomic layer and with the film lattice constant a:

$\Rightarrow \Delta G_{2D}=-\pi r^2a\Omega^{-1}\Delta\mu+\pi r^2a\Omega^{-1/3}(\gamma_F+\gamma_{S/F}-\gamma_S)+2\pi r\gamma_E$

Critical cluster sizes (number of atoms, $j_{cr}$)

3D growth: $\Delta G_{3D}(J_{cr})=\frac{4}{27}\frac{X^3}{\Delta \mu^2}, j_{cr}=(\frac{2X}{3\Delta\mu})^3$

2D growth: $\Delta G_{2D}(J_{cr})=\frac{1}{4}\frac{Y}{\Delta \mu^2}, j_{cr}=(\frac{Y}{2\Delta\overline{\mu}})^2$

For 2D growth the second term is also negative, so augments supersaturation.

$\Delta\overline{\mu}$ is an effective quantity: $\Delta\overline\mu=\delta\mu-(\gamma_F-\gamma_{S/F}-\gamma_S)\Omega^{2/3}$ (which determines the critical 2D cluster size)

$\Rightarrow$ 3D cluster growth for $\Delta\mu>0$, supersaturation; 2D growth for $\Delta\overline\mu>0$ ($\Delta\mu \leq 0$), subsaturation

The rate $J_N$ at which critical nuclei are formed:

$$J_N=K\exp[-\Delta G(j_{cr}/kT)]$$

$J_N \rightarrow$ the growth rate of a film

increasing $\zeta$ $\rightarrow$ smaller aggregation sizes & higher numbers of nuclei

In vacuum deposition, high supersaturation at low substrate temp, the critical size of a nucleus might be very small - it may contain only a few atoms. In this limit then, the applicability of the classical theory with its macroscopically determined terms $\gamma_F, \gamma_S, \gamma_{S/F}$ is questionable.

## Kinematic theory of surface scattering

**Wave-like projectiles scattering**

(1) Interaction of the particle wit solid - in particular with its surface - is described by a potential:

$$V(r, t)=\sum_nv[r-\rho_n(t)]$$

- $\bf{n}$(=triple m,n,p) labels atoms in a primitive lattice
- $v$: interaction potential with a single near-surface atom (substrate or adsorbate) at the instantaneous position $\rho_n(t)\cdot\rho_n(t)$ contains the time-independent equilibrium position $r_n$ and displacement $s_n(t)$ from equilibrium: $\rho_n(t)=r_n+s_n(t)$

(2) Monoenergetic particles with energy E are incident with a wave vector $k(E=\hbar^2k^2/2m)$ and scattered into a state $k'$. Scattering probability per unit time from $k$ into $k'$:

$$W_{kk'}=\lim_{\tau\rightarrow{\infty}}\frac{1}{\tau}|c_{kk'}(\tau)|^2$$

with the transition amplitude:

$$c_{kk'}=\frac{-i}{\hbar}\int dr\int_0^{\tau}dt\psi_S^*(r,t)V(r)\psi_i(r,t)$$

Assume a plane wave character:

$$\psi_i(r,t)=V^{-1/2}e^{i(k\cdot r-Et/\hbar)}$$

$$\psi_S(r,t)=V^{-1/2}e^{i(k'\cdot r-E't/\hbar)}$$

Probability amp for scattering within a time $\tau$:

$$c_{kk'}(\tau)=\frac{-i}{\hbar}\sum_n\int_0^{\tau}e^{i(E'-E)t/\hbar}\int e^{i(k-k')\cdot r}[r-\rho_n(t)]drdt$$


vector $r-\rho_n(t)=\xi$: a position of a moving atom, whose nucleus is at the instantaneous position $\rho_n(t)$, with $dr=d\xi$

$$c_{kk'(\tau)}=\frac{-i}{\hbar}\sum_n\int_0^\tau dt e^{i(E'-E)t/\hbar}e^{i(k-k')\cdot \rho_n(t)}\int d\xi v(\xi)e^{-i(k-k')\cdot\xi}$$

atomic scattering factor: $f(K)=\int d\xi v(\xi)e^{-iK\cdot\xi}$

instantaneous position of an atom near the surface:

$$\rho_n(t)=r_n+s_n(t)=r_{n\parallel}+z_p\hat{e}_\perp+s_n(t)$$

$[n=(n\parallel, p)]$ $\rightarrow$ the translational symmetry parallel to the surface which allows a Fourier representation:

$$\rho_n(t)=r_{n\parallel}+z_p\hat{e}_\perp+\sum_{q\parallel}\hat{s}(q_\parallel,z_P)\exp[\pm iq_\parallel\cdot r_{n\parallel}]\pm i\omega(q_\parallel)t$$

the displacement from equilibrium $s_n(t)$ described by a harmonic wave with freq $\omega(q_\parallel)$ parallel to the surface.

the most general excitation: superposition of modes with different $\omega(q_\parallel)$

$$c_{kk'}(\tau)=\frac{-i}{\hbar}f(k)\sum_{n_\parallel,P}\int_0^\tau dt e^{i(E'-E)t/\hbar}\times\exp\{-iK\cdot[r_{n\parallel}+z_P\hat{e}_\perp+\sum_{q\parallel}+\sum_{q_\parallel}\hat{s}(q_\parallel,z_P)\exp(\pm iq_\parallel\cdot r_{n\parallel}\pm i\omega(q_\parallel)t)]\}$$

displacement from equilibrium are small, the exponential func expanded:

$$exp\{-iK\cdot[\sum_{q\parallel}\hat{s}(q_\parallel,z_P)e^{\pm i\cdots}]\} \\ \approx 1-iK\cdot[\sum_{q_\parallel}\hat{s}(q_\parallel,z_P)\exp[\pm i(q_\parallel\cdot r_{n_\parallel}+\omega t)]]+\cdots$$

Only consider the first two terms in the expansion:

$$c_{kk'}(\tau)=\frac{-i}{\hbar}f(k)\sum_{n_\parallel,P}\int_0^\tau dt \exp{i(E'-E)t/\hbar}\exp[-iK\cdot(r_{n\parallel+z_P}\hat{e}_\perp)]\times \{1-i\sum_{q_\parallel}K\cdot\hat{s}(q_\parallel,z_P)\exp[\pm i(q_\parallel\cdot r_{n_\parallel}+\omega t)]\}$$

First 'elastic' term not contain the vibrational amp $\hat{s}$, elastic scattering probability, the time limit $\tau\rightarrow\infty$ $\Rightarrow$ $\delta(E'-E)$; summation over 2D set $n_\parallel=(m,n)$ runs over the whole surface to infinity; it contains terms of the form:

$$\sum_{m,n}e^{iK\cdot(ma+nb)}=\sum_{m,n}(e^{-iK\cdot a})^m(e^{-iK\cdot b})^n$$

an evaluation via a geometrical series gives, in the limit $m,n\rightarrow\infty$, a nonvanishing contribution only for:

$$K\cdot a = 2\pi h, K\cdot b = 2\pi k; h,k \text{interger}$$

With $K=K_{\parallel}+K_\perp\hat{e}_\perp$

fulfilled when $K_\parallel = k_\parallel'-k_\parallel=G_\parallel$

$\Rightarrow$ 2D analogs of the three Laue equations for the bulk scattering of X-rays.

(3) Elastic scattering probability:

With N being the number of surface atoms

$$W_{k'k}^{(el)}=\frac{2\pi}{\hbar}N|f(k)\sum_P\exp(iK_\perp z_P)|^2\delta(E'-E\delta_{K_\parallel,G_\parallel})$$

$z_p=pc$, a summation of p up to infinity $\rightarrow$ third Laue condition.

(4) Inelastic scattering probability:

the second terms, the inelastic scattering amplitude:

$$c_{kk'}^{inel}(\tau)=\frac{-i}{\hbar}f(K)\sum_{n_\parallel,p}\int_0^\tau dt\exp[i(E'-E)t/\hbar]\exp[-iK\cdot(r_{n_\parallel}+z_p\hat{e}_\perp)]\times(-i)\sum_{q_\parallel}K\cdot\hat{s}(q_\parallel,z_p)\exp[\pm i(q_\parallel\cdot r_{n_\parallel}+\omega t)] \\ = -\hbar^{-1}f(K)\sum_{n_\parallel,p,q_\parallel}K\cdot\hat{s}(q_\parallel,z_p)\exp[-i(K_\parallel\mp q_\parallel)\cdot r_{n_\parallel}] \times \exp(-iK_\perp z_p)\int_0^\tau[i(E'-E\pm\hbar\omega)t/\hbar]dt$$

inelastic scattering probability:

$$W_{kk'}^{inel}=\frac{2\pi}{\hbar}N\sum_{q_\parallel}\delta(E'-E\pm\hbar\omega(q_\parallel))\delta_{K_\parallel\pm q_\parallel,G_\parallel}\times|f(K)\sum_p K\cdot\hat{s}(q_\parallel,z_p)\exp(-iK_\perp z_p)|^2$$

delta function and Kronecker symbol ensure energy conservation $E'=E\mp\hbar\omega(q_\parallel)$

the conservation of the wave-vector component parallel to the surface ($K_\parallel=k_\parallel'-k_\parallel$)

$$k_\parallel'=k_\parallel\pm q_\parallel + G_\parallel$$

$\Rightarrow$ <font color='blue'> a straightforward consequence of 2D translational symmetry within the surface <font>

in all inelastic surface scattering processes the energy lost (or gained) in scattering must to found as the quantum energy $\hbar\omega(q_\parallel)$ of an excited surface mode (vibrational). An extension of the present formalism can be made to include electronic excitations of the surface.
    
second factor $\rightarrow$ a selection rule for surface scattering
    
$\Rightarrow$ symmetry of an excited vibrational (or electronic) surface excitation: inelastic scattering probability vanishes if $K$ is perpendicular to $\hat{s}(q_\parallel, z_P)$. $\hat{s}$ are Fourier components of the atomic vibration and so have direction of the atomic displacement (or the electronic dipole moment).
    
Only see those excitation for which the atomic displacement has a component parallel to the scattering vector $K=k'-k$. For scattering in the specular direction only modes with vibrational direction normal to the surface are detectable. Vibrations parallel to the surface can only be studied if one observes scattering with off-specular geometry. In calculating $K=k′−k$, consider that due to the inelastic process $|k'|$ is different from $|k|$, even for specular detection, there is a small component of K parallel to the surface. The size of this component depends on the ratio $\hbar\omega/E$

## Step fluctuations

**Energy of a fluctuating step**

The total energy of a surface configuration (depend only on the height difference $|h_{ij}-h_{i+n,j+m}|$ at sites $(i,j)$ and $(i+n,j+m)$):

$$H=J\sum_{i,j,m,n}|h_{ij}-h_{i+n,j+m}|^p$$

- energy is normalized with respect to the energy of the flat surface
- In SOS model, $H$ is linear function of the energy differences ($p=1$) $\simeq$ Hamiltonian of a 2D spin distribution as described by the Ising model (nearest-neighbor interactions); $J$ corresponds to twice the coupling stength b
- In _discrete Gaussian model_, H depends quadratically on the energy difference ($p=2$)

(1) SOS model:

- discrete description:

In case of a step on the surface of length $Na_\parallel$ ($a_\parallel$ is the atomic unit parallel to the step):

$$H=J\sum_{j=1}^N\frac{|x_j-x_{j+1}|}{a_\perp}$$

- continuum approach:

based on the continuum model for the kinetics of faceting and surface relaxation by Mullins. 2D problem reduced to 1D case of a fluctuating step profile. 

the total free energy of a fluctuating step can be expressed in terms of the line tension $\gamma$, which is the specific free step energy (per unit length):

$$f=\int\gamma ds = \int\gamma(\frac{\partial x}{\partial y})\sqrt{1+(\frac{\partial x}{\partial y})^2}dy$$

$\sqrt{1+(\frac{\partial x}{\partial y})^2}$: the small step segment $ds$ along the step profile

Maryland notation: x is the coordinate perpendicular to the mean orientation of the step and y is the coordinate parallel to it.

Variation of the step position $x(y)$ is accompanied by a variation in the total step length $\rightarrow$ the step free energy $f$

chemical potential $\mu$ $\leftarrow$ changes the number of atoms by $dN=\delta x\cdot dy/\Omega$

$$\delta f = \frac{1}{\Omega}\int(\delta x \mu)dy$$

with the chemical potential:

$$\mu = -\Omega\frac{\partial}{\partial y}\frac{\partial}{\partial(\partial x/\partial y)}(\gamma(\frac{\partial x}{\partial y})\sqrt{1+(\frac{\partial x}{\partial y})^2})$$

$\partial x/\partial y = \tan\theta$, local curvature $K=-\partial^2x/\partial y^2[1+(\partial x/\partial y)^2]^{-3/2}$

$\Rightarrow$ $\mu = \Omega K(\gamma+\frac{\partial^2\gamma}{\partial\theta^2})$ 

For LT, $K\approx-\partial^2x/\partial y^2$

$\Rightarrow$ $\mu=-a_\perp a_\parallel\frac{\partial^2x}{\partial y^2}(\gamma+\frac{\partial^2\gamma}{\partial\theta^2})$

step-edge stiffness: $\tilde{\gamma}=\gamma+\frac{\partial^2\gamma}{\partial\theta^2}$

(a) SOS model:

$\tilde{\gamma}=2k_BTa_\parallel/a_\perp^2\sinh(J/2k_BT)$

(b) Ising model

$\tilde{\gamma}=k_BTa_\parallel/a_\perp^2\sinh(J/k_BT)$

**Langevin equation**

The gradient in the chemical potential $\mu$: Non-equilibrium $\rightarrow$ equilibrium

the kinetics of the dominant atomic processes (mediate the mass transport) $\rightarrow$ relaxation rate

1D system of step fluctuations, the thermodynamics described by a _Langevin equation_:

$$\frac{\partial x(y,t)}{\partial t}=\frac{\Gamma_h}{k_BT}\Lambda(x,t)+\eta(x,t)$$

- $\Lambda(x,t)$: a to-be-determined function of $x$ and $t$ (depends on dominant mass transport, expressed by $\mu(x,t)$) and diffusion kernel $P(l)$
- $\eta(x,t)$: a noise term
- $\Gamma_h = a_\perp^2a_\parallel/\tau_h$: _friction function_ (depends on the sticking probability of an atom at a distinct step site, so a func of the activation barriers relevant for the dominant mass transport processes)
- $\tau_h$: the time between two detachment processes from a step site

attachment/detachment from step sites onto terrace sites, the jump rate:

$$\Gamma_{EC}=a_\perp^2a_\parallel ve^{-E_{ad}/k_BT}(e^{-E_{st}^{low}+e^{-E_{st}^{up}/k_BT}})$$

- $E_{st}^up$: Ehrlich-Schwoebel (ES) barrie or step-edge barrier

$$\gamma(x,t)=\int_0^{L/2}P(l)\frac{\partial^2u}{\partial x^2}dl$$
- L: length of the step
- P(l): describe probability that an atom at site A reaches site B at distance $l$

When the step fluctuations are mediated by edge diffusion, the diffusion kernel is given by the probability that an atom at the step edge in one atomic distance to the site $l$ reaches site $l$:

$$P(l)=\frac{1}{2}[\delta(l-a_\parallel)+\delta(l+a_\parallel)]$$

In the case of evaporation/condensation, the atom detaches from the step site onto the terrace site and surface diffusion is fast.

the correlation length $\xi$ between atomic sites >> the step length L.

the diffusion kernel is constant:

$$P(l)=L^{-1}$$

In the case of slow terrace diffusion, the situation is more complex.

### Coarsening of monolayer island structures

**Ostwald ripening of islands**

Lifshitz, Slyozov and Wagner: classical theory of Ostwald ripening for 3D particles in colloidal liquids

Chakraverty, Wynblatt, Gjostein: derive for 2D problem of islands on surfaces

The classical Ostwald theory of island ripening is formulated for an individual circular island with radius $r_i$ among other islands. The island is surrounded by a circular (2D) sphere with constant radius in which the adatom concentration is a mean adatom concentration field formed by the other islands.

Island configurations with fixed boundary conditions, which may be realized in experiment and solved analytically under certain assumptions.
- adatom island located in the center of a larger circular vacancy island
- vacancy island within an adatom island
- vacancy island within a vacancy island
- adatom island on top of another adatom island

chemical potential of a circular island: (Gibbs-Thomson relation)

$$\mu(r)=\Omega\frac{\gamma}{r}$$

- $\gamma$: the step line tension (= step-edge stiffness $\tilde{\gamma}$)
- $\Omega$: the area of an atom

PS: high temp limit of $\mu=a_\parallel a_\perp\tilde{\gamma}(\theta=0)\frac{\partial^2y}{\partial x^2}\lvert_{(x=0)}$

holds also fro hexagonal and quadratic islands but a circular shape

Intro a _shape factor_ $\eta$: relate an island with side length $s$ to an equivalent circular island with radius $R$ with the same area:

$$\mu(r)=\eta\Omega\frac{\gamma}{R},\eta=\sqrt{\frac{2\sqrt{3}}{\pi}}(\text{hex}), \eta=\frac{2}{\sqrt{\pi}}(\text{sq})$$

PS: The atomic area is given by $\Omega=a_\parallel/a_\perp$ for fcc metal surfaces, $a_\parallel$ and $a_\perp$ being the distance between atomic rows along and perpendicular to the dense $\langle110\rangle$-direction

The gradient in $\mu$ between the edge of the innear and the outer island radius $\rightarrow$ a mass current the islands

The number of atoms per time unit attaching to the perimeter of the innear island:

$$j_{\text{att}}=-\kappa\rho(r_i)$$

- $\kappa$: rate const
- $\rho(r_i)$: concentration of adatoms per atom on the terrace in close proximity to the innear island radius $r_i$

In equilibrium:

$$-j_{\text{det}}\stackrel{!}{=}-\kappa\rho_{\text{eq}}(r_i)=j_\text{att}$$

$$j_\text{net}=j_\text{det}+j_\text{att}=\kappa(\rho_{\text{eq}}(r_i)-\rho(r_i))$$

The net current flow must match the flow by terrace diffusion $j_\text{diff}$:

$$j_\text{net}\stackrel{!}{=}j_\text{diff}=-\nu_\text{diff}a\text{grad}\rho(r_i)$$

- $\nu_\text{diff}=\nu_0\exp(-E_\text{diff}/k_BT)$: hopping rate on the terrrace
- a: atomic units (related to the atomic distance between equivalent lattice positions)

PS: a = the nearest-neighbor distance $a_\parallel$ (sq); 3-fold fcc hollow sites (hex)

$\kappa=sv_{\text{diff}}$: the product of the hopping rate on the terrace and an additional sticking rate $s$ at the island perimeter

sticking rate: $v_{\text{stick}}=v_0^*\exp(-(E_{\text{stick}}+E_{\text{diff}})/k_BT)$
- $s=1$ $\rightarrow$ no additional sticking barrier is present
- $s=0$ $\rightarrow$ the barrier is infinitely large

A differential equ for the adatom concentration:

$$\rho(r_i)-\frac{a}{c}\text{grad}\rho(r_i)=\rho_{\text{eq}}(r_i)$$

The adatom concentrations obeys the equ for stationary diffusion which is the Laplace eq., the solution:

$$\rho(r_i)=\ln|\frac{r_i}{r_0}|\frac{\rho_{\text{eq}}(r_i)-\rho_\text{eq}(r_o)}{\ln|\frac{r_i}{r_o}|-\frac{a}{sr_i}}+\rho_\text{eq}(r_o)$$

<font color = 'blue'>Note: for a vacancy island, the curvature of the island perimeter is negative. Hence, the radius ro at the outer island perimeter is interpreted as negative variable. In the case of an adatom island on top of another adatom island, the outer island perimeter is positive and for the problem of a vacancy island within a vacancy island both radii are negative.<font>
    
$\Rightarrow$ $j_\text{net}=v_\text{diff}\frac{a}{r_i}\frac{\rho_\text{eq}(r_i)-\rho_\text{eq}(r_o)}{\ln|\frac{r_o}{r_i}|+\frac{a}{sr_i}}$
    
The reduction rate of particles $N$ in the island with time:
    
$$\frac{dN}{dt}=-j_\text{net}\frac{2\pi r_i}{a}$$
    
$$\frac{dN}{dt}=-2\pi v_\text{diff}\frac{\rho_\text{eq}(r_i)-rho_\text{eq}(r_o)}{\ln|\frac{r_o}{r_i}|+\frac{a}{sr_i}}$$
    
equilibrium concentration: $\rho_\text{eq}(r_i)=\rho_\infty\exp(\mu(r_i)/k_BT)$ with $\mu=\eta\gamma\Omega/r_i$ and $\rho_\infty=\exp(-E_\text{ad}/k_BT)$ the equilibrium adatom concentration on the terrace.
    
$$\rho_{eq}(r_i)=\exp\{-\frac{E_\text{ad}}{k_BT}\}\exp\{\frac{\eta\gamma\Omega}{r_ik_BT}\}$$
    
$$\frac{dA}{dt}=-2\pi v_0\Omega\exp\{-\frac{E_\text{ad}+E_\text{diff}}{k_BT}\}\frac{\exp\{\frac{\eta\gamma\Omega}{k_BT}(\frac{1}{r_i}-\frac{1}{r_o})\}}{\ln|\frac{r_o}{r_i}|+\frac{a}{sr_i}}$$

approx for 2 limiting cases:

the detachment of atoms at the island edge is fast compared to the diffusion on the terrace.
    
$a/r_i\ll\ln|r_o/r_i|$ and with the additional, not necessarily fulfilled approx $\exp(\eta\gamma\Omega/r_{i,o})\approx\eta\gamma\Omega/r_{i,o}$
    
$\Rightarrow$
    
$$\frac{dA}{dt}\approx-\frac{2\pi v_o\Omega}{\ln|\frac{r_o}{r_i}|}\exp\{-\frac{E_\text{ad}+E_\text{diff}}{k_BT}\}\frac{\eta\gamma\Omega}{k_BT}(\frac{1}{r_i}-\frac{1}{r_o})$$

## Wettability in Fe3Sn/Pt(111) Growth

**Wettability** describes how well a material (Fe₃Sn) spreads or forms a uniform layer on a substrate (Pt(111)). It is controlled by **surface energy, adsorption energy, and interface interactions**.

---

## **1. Mathematical Definition: Spreading Parameter (\(S\))**
The **spreading parameter** (\(S\)) determines whether Fe₃Sn forms a **continuous monolayer** or **islands** on Pt(111):

\[
S = E_{\text{Pt(111)}} + E_{\text{Fe}_3\text{Sn,free}} - E_{\text{Fe}_3\text{Sn/Pt(111)}}
\]

Where:
- \(E_{\text{Pt(111)}}\) = Energy of clean Pt(111).
- \(E_{\text{Fe}_3\text{Sn,free}}\) = Energy of a **free-standing Fe₃Sn** monolayer.
- \(E_{\text{Fe}_3\text{Sn/Pt(111)}}\) = Energy of **Fe₃Sn adsorbed on Pt(111)**.

---

### **2. Interpretation of \(S\)**
- **\(S > 0\) (Positive Spreading Parameter)**
  - **Fe₃Sn prefers to form a monolayer** (wetting layer).
  - This means **Fe₃Sn spreads well** over Pt(111).
  - Layer-by-layer (2D) growth is likely.

- **\(S < 0\) (Negative Spreading Parameter)**
  - **Fe₃Sn prefers to form 3D islands (Stranski-Krastanov growth).**
  - This means Fe₃Sn **does not wet** Pt(111) well.
  - **Island formation is favored** instead of a smooth film.

---

## **3. Wettability and Fe₃Sn Growth Mode on Pt(111)**
The wettability of Fe₃Sn on Pt(111) determines which **growth mode** occurs:

| **Growth Mode** | **Condition** | **Expected Film Structure** |
|---------------|--------------|----------------------------|
| **Layer-by-layer (2D growth)** | \( S > 0 \) | Fe₃Sn forms a **uniform monolayer** |
| **Island growth (3D growth)** | \( S < 0 \) | Fe₃Sn forms **discrete islands** |
| **Layer + islands (SK growth)** | \( S \approx 0 \) | Fe₃Sn starts as a monolayer, then islands grow |

## Interface Energy ($E_{interface}$)

Quantifies energy cost at the Fe3Sn/Pt(111) boundary. Important for heterostructure formation and wetting behavior.
### **Interface Energy (\(E_{\text{interface}}\))**  
The **interface energy** quantifies the stability and strength of bonding between Fe₃Sn and the Pt(111) substrate. It helps determine whether Fe₃Sn will **form a stable interface**, dissolve into Pt, or segregate.

---

## **1. Mathematical Definition**
The interface energy per unit area is calculated as:

\[
E_{\text{interface}} = \frac{E_{\text{Fe}_3\text{Sn/Pt(111)}} - (E_{\text{Pt(111)}} + E_{\text{Fe}_3\text{Sn,bulk}})}{A}
\]

where:
- \(E_{\text{Fe}_3\text{Sn/Pt(111)}}\) = Total energy of Fe₃Sn adsorbed on Pt(111).
- \(E_{\text{Pt(111)}}\) = Energy of the clean Pt(111) substrate.
- \(E_{\text{Fe}_3\text{Sn,bulk}}\) = Energy of **bulk Fe₃Sn**.
- \(A\) = Interface area (in Å² or nm²).

---

## **2. Interpretation of \(E_{\text{interface}}\)**
- **\(E_{\text{interface}} < 0\) (Negative Interface Energy)**
  - Fe₃Sn **forms a stable interface** with Pt(111).
  - Strong Fe₃Sn-Pt bonding → Favorable wetting and adhesion.

- **\(E_{\text{interface}} > 0\) (Positive Interface Energy)**
  - The interface is **unfavorable**.
  - Fe₃Sn may **segregate or dissolve** into Pt instead of forming a continuous film.

---

## **3. Why is \(E_{\text{interface}}\) Important for Fe₃Sn/Pt(111)?**
- **If \(E_{\text{interface}}\) is low (negative),** Fe₃Sn will **bind well to Pt(111)**.
- **If \(E_{\text{interface}}\) is high (positive),** Fe₃Sn will **prefer to form islands or separate phases** instead of a uniform film.
- Helps decide **whether Pt atoms should be added to Fe₃Sn** to improve interface stability.
- Important for **electronic structure changes at the interface**.

---

## Diffusion Barrier Energy ($E_{diff}$)

Determines whether Fe and Sn atoms diffuse into Pt(111) instead of forming Fe3Sn.

### **Diffusion Barrier Energy (\(E_{\text{diff}}\))**  
The **diffusion barrier energy** quantifies how much energy is required for an Fe or Sn atom to move across or into the Pt(111) surface. It helps determine:

- **Whether Fe and Sn will remain on the surface to form Fe₃Sn**
- **Or if they will dissolve into the Pt(111) substrate**

---

## **1. Mathematical Definition**
Diffusion barrier energy is calculated using the **Nudged Elastic Band (NEB) method**, which finds the minimum energy path (MEP) for an atom moving from one site to another.

\[
E_{\text{diff}} = E_{\text{TS}} - E_{\text{initial}}
\]

where:
- \(E_{\text{TS}}\) = Energy of the **transition state** (highest energy along the diffusion path).
- \(E_{\text{initial}}\) = Energy of the **initial adsorption site**.

This represents the **activation energy required for an atom to move**.

---

## **2. Why is \(E_{\text{diff}}\) Important for Fe₃Sn on Pt(111)?**
- If **\(E_{\text{diff}}\) is high (\(> 1.0\) eV)** → Fe and Sn are unlikely to diffuse **into** Pt, allowing Fe₃Sn to form a **stable monolayer**.
- If **\(E_{\text{diff}}\) is low (\(< 0.5\) eV)** → Fe and Sn atoms can **dissolve into Pt(111)**, preventing Fe₃Sn monolayer formation.
- Helps determine **growth conditions** (temperature, annealing) to optimize Fe₃Sn stability.

---

## **3. How to Compute \(E_{\text{diff}}\) using DFT (VASP)?**
To calculate diffusion barriers, we use the **Nudged Elastic Band (NEB) method** in VASP.

### **Step 1: Define the Initial and Final Positions**
- Place **Fe or Sn** at an **initial adsorption site** on Pt(111).
- Define the **final position** (e.g., another site on Pt(111) or inside the substrate).

### **Step 2: Generate Intermediate Images**
- Use **5–10 intermediate positions** between the initial and final sites.
- These images represent **atomic movement** during diffusion.

### **Step 3: Run NEB Calculation**
- Perform **VASP NEB calculations** to find the **energy profile**.
- Extract the **maximum energy point (transition state, \(E_{\text{TS}}\))**.

### **Step 4: Compute \(E_{\text{diff}}\)**
- Use:

\[
E_{\text{diff}} = E_{\text{TS}} - E_{\text{initial}}
\]

## Surface Free Energy

The **surface free energy** ($\gamma$) is a fundamental property that quantifies the energetic cost of creating a new surface in a material. It is defined as the excess free energy per unit area associated with the surface of a solid or liquid compared to the bulk.

### Mathematical Definition

The surface free energy is given by: $$\gamma = \frac{\partial G}{\partial A}$$

where:
- $G$ is the Gibbs free energy,
- $A$ is the surface area.

For a solid, the surface free energy arises due to the unsaturated bonds of surface atoms, which are not fully coordinated as in the bulk.

### Units

Surface free energy is typically measured in:
- Joules per square meter (J/m²)
- Ergs per square centimeter (erg/cm²)
- Millinewtons per meter (mN/m), often used for liquids

### Surface Free Energy in Different Systems
1. **Metals and Alloys**  
   - Typically range between **1–3 J/m²**.
   - Strongly depends on the crystallographic orientation.
   - FCC metals like Pt, Au, and Cu have lower surface energies compared to BCC and HCP metals.
   
2. **Semiconductors**  
   - Surface reconstructions and passivation can significantly alter the surface free energy.
   - Example: Si(100) surface has a lower energy than Si(111) due to dimerization.

3. **Liquids**  
   - Equivalent to **surface tension**, denoted as \( \gamma \).
   - Water: **72.8 mN/m at 25°C**.
   - Mercury: **485 mN/m** (high due to strong metallic bonding).

4. **Thin Films and Surfaces on Substrates**  
   - When deposited on a substrate, the surface energy can be modified due to interactions with the substrate.
   - This affects **wetting, adhesion, and film growth** mechanisms.

### **Relevance to Pt(111) and Fe₃Sn Growth**
- The **surface free energy of Pt(111)** is around **2.5 J/m²**, making it relatively stable.
- When **Sn and Fe** are deposited on Pt(111), the surface free energy influences **wetting, diffusion, and alloy formation**.
- The competition between **surface energy minimization and strain effects** determines whether Fe₃Sn will grow as an ordered layer or form islands.

## Raman effect

(inelastic scattering of photons in the visible or UV spectral range) allows one to study the vibrational properties of thin films

Due to the lack of translational symmetry in amorphous material, the wave vector is not conserved upon light scattering and phonons from all over the Brillouin zone contribute to give essentially an image of the phonon density (versus energy) rather than of single sharp phonon branches.

## Cold Atom Experiments: Exploring Quantum Phenomena at Ultra-Low Temperature

Cooling neutral atoms or ions to near absolute zero temp (typically in the nanoKelvin to microKelvin range) to explore fundamental quantum mechanical phenomena. unprecedented control over atomic interactions, trapping potentials, and qauntum states.

### **Cold Atom Experiments: Exploring Quantum Phenomena at Ultra-Low Temperatures**

**Cold atom experiments** involve cooling **neutral atoms** or **ions** to **near absolute zero temperatures** (typically in the **nanoKelvin** to **microKelvin** range) to explore fundamental quantum mechanical phenomena. These systems offer **unprecedented control** over atomic interactions, trapping potentials, and quantum states, making them powerful tools for studying **quantum mechanics, condensed matter physics, and quantum information**.

---

### **Key Concepts in Cold Atom Experiments**

#### **1. Laser Cooling and Magneto-Optical Traps (MOT)**
- **Laser Cooling**:
  - Uses **Doppler cooling** and **Sisyphus cooling** to reduce atomic velocities.
  - Achieves temperatures close to the **Doppler limit** (~100 μK for alkali atoms).
- **Magneto-Optical Trap (MOT)**:
  - Combines laser cooling with **magnetic field gradients**.
  - Captures and cools atoms to **microKelvin** temperatures.

#### **2. Evaporative Cooling and Bose-Einstein Condensation (BEC)**
- **Evaporative Cooling**:
  - Lowers temperature by selectively removing high-energy atoms.
  - Essential for reaching **nanoKelvin** temperatures.
- **Bose-Einstein Condensation (BEC)**:
  - Achieved with **bosonic atoms** (e.g., Rubidium-87, Sodium-23).
  - At ultra-low temperatures, atoms occupy the **same quantum state**, forming a **macroscopic quantum object**.

#### **3. Optical Lattices and Quantum Simulators**
- **Optical Lattices**:
  - Periodic potential created by **interfering laser beams**.
  - Mimics the **crystal lattice** in solids, allowing exploration of **Hubbard models, superfluid-Mott transitions**, and **topological states**.
- **Quantum Simulators**:
  - Cold atoms in tailored potentials simulate **condensed matter systems**.
  - Examples include simulating the **Hubbard model**, **Lieb lattice**, and **spin chains**.

#### **4. Feshbach Resonances and Tunable Interactions**
- Use of **magnetic fields** to tune **interatomic interactions**.
- Enables exploration of the **BCS-BEC crossover**, **Efimov states**, and **strongly correlated systems**.

#### **5. Ion Traps and Quantum Information**
- **Paul and Penning traps** confine ions for **quantum computing**.
- Use **laser cooling** to achieve quantum ground states.
- Implement **quantum gates** and **entanglement** for **quantum information processing**.

---

### **Key Applications of Cold Atom Experiments**
✅ **Quantum Simulations**: Study complex many-body systems like **high-temperature superconductors** and **quantum magnetism**.  
✅ **Precision Measurements**: Atomic clocks, **fundamental constant measurements**, and **gravitational wave detection**.  
✅ **Quantum Information**: Quantum computation with trapped ions and cold atoms.  
✅ **Exploring Exotic States**: BEC, **Tonks-Girardeau gas**, **topological phases**, and **synthetic gauge fields**.  
✅ **Testing Fundamental Physics**: Investigate **quantum entanglement**, **non-locality**, and **Bell's inequalities**.  

---

### **Examples of Cold Atom Experiments**
1. **BEC in Rubidium-87 and Sodium-23**:  
   - First BEC observed in 1995, leading to a **Nobel Prize** in 2001.  

2. **Simulating Hubbard Models**:  
   - Use of **optical lattices** to explore **Mott insulators** and **superfluid-Mott transitions**.  

3. **Topological Matter**:  
   - Cold atoms in synthetic lattices demonstrating **Chern insulators** and **quantum spin Hall effects**.  

4. **Atomic Clocks**:  
   - Optical lattice clocks with precision up to **18 decimal places**, used for redefining the **second**.  

---

### **Conclusion**
**Cold atom experiments** allow us to **probe and control quantum phenomena** with high precision, bridging **atomic physics**, **condensed matter physics**, and **quantum information**. They are invaluable for exploring **new states of matter** and **testing fundamental theories**.
